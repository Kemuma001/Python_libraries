```python
import matplotlib.pyplot as plt 
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Simple line graph")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
plt.plot(x,y, marker='o',linestyle='--')
plt.title("styled Line Graph")
plt.show()
```

```python
plt.plot(x,y, 'r--o')
plt.show()
```

```python
z=[4,5,6,7,8]
w=[1,2,3,4,5]
plt.plot(z,w, 'b--o')
plt.title("Multiple line graph")
plt.xlabel("Z Values")
plt.ylabel("W Values")
plt.show()

```

```python
categories=["Nairobi","Mombasa","Kisumu","Eldoret"]
prices=[4500,3800,3000,2500]

plt.bar(categories,prices)
plt.title("Average Airbnb prices by city")
plt.xlabel("City")
plt.ylabel("Price")
plt.show()
```

```python
#we use barh for horizontal bars
plt.barh(categories,prices)
plt.title("Average Airbnb prices by city")
plt.show()
```

```python
#Scatter plot, if the dots form straight line mean strong relationship while scattered means weak relationship
size=[20,30,40,50,60]
price=[2000,2500,3000,3500,4000]

plt.scatter(size,price)
plt.title("House siize vs price")
plt.xlabel("size")
plt.ylabel("price")
plt.show()
```

```python
#Histogram, it shows the distribution of data, how often values appear, whether data is skewed or its normal
prices=[2000,2200,2500,3000,3200,3500,4000,4500] 
plt.hist(prices, bins=5)
plt.title("Price distribution")
plt.xlabel("price")
plt.ylabel("Frequency")
plt.show()
```

```python
#understanding figure
#figsize(8,5) means width 8 and height 5
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.figure(figsize=(8,5))
plt.plot(x,y)
plt.title("bigger Graph")
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[1,3,5,7,9]

plt.plot(x,y1)
plt.plot(x,y2)

plt.title("Two line graphs")
plt.show()
```

```python
#Adding legends, plt.legend() tells matplotlib to use the labels provided 

plt.plot(x,y1, label="Even numbers")
plt.plot(x,y2, label="Odd numbers")

plt.legend()
plt.title("Two lines with legend")
plt.show()
```

```python
#Adding the grid, makes it easier to read the graph and identify values
plt.plot(x,y1)
plt.grid(True)
plt.show()
```

```python
#Saving the graph as image
plt.plot(x,y1)
plt.title("Saved Graph")
plt.savefig("my_graph.png")
plt.show()
```

```python
x=[1,2,3,4,5]
y2=[4,6,8,10,12]
y3=[1,3,5,7,9]

plt.plot(x,y2, label="Even numbers")
plt.plot(x,y3, label="Odd numbers")
plt.title("Two lines with legend")

plt.legend()
plt.figure(figsize=(10,5))
plt.savefig("my_graph.png")

plt.grid(True)
plt.show()
```

```python
#Understanding object oriented style
plt.plot(x,y1)
plt.title("My First Plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]

fig, ax=plt.subplots()
ax.plot(x,y)
ax.set_title("Object oriented plot")
ax.set_xlabel("X values")
ax.set_ylabel("Y values")

plt.show()
```

```python
#Multiple plots in one figure
x=[1,2,3,4,5,]
y1=[2,4,6,8,10]
y2=[1,3,5,7,9]

fig, ax =plt.subplots(1,2,figsize=(10,4)) 
ax[0].plot(x,y1)
ax[0].set_title("Even numbers")

ax[1].plot(x,y2)
ax[1].set_title("odd numbers")

plt.show()
```

```python
#Customazing color properly
x = [1,2,3,4,5]
y1 = [2,4,6,8,10]

fig, ax = plt.subplots()
ax.plot(x, y1, color="green", marker="o", markersize=8)
plt.show()
```

```python
#Scatter plot  with transparency (Alpha)
import numpy as np
 
x=np.random.rand(50)
y=np.random.rand(50)

plt.scatter(x,y, alpha=0.5)
plt.title("Scatter with transparency")
plt.show()
```

```python
#Adding text to a plot 
plt.plot(x,y1)
plt.text(2,6, "Important point")
plt.show()
```

```python
#Adding a trend line to scatter plot
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([2,4,5,4,5])

plt.scatter(x,y)
m,b =np.polyfit(x, y, 1)
plt.plot(x,m*x+b,) 
plt.title("Scatter with trend line")
plt.show()
```

```python
#Bar chart
booking=[120,80,45]
cities=["Nairobi","Mombasa","Kisumu"]

plt.bar(cities,booking)
plt.xlabel("City")
plt.ylabel("Number of bookings")
plt.title("Bookings per city")
plt.show()
```

```python
#Histogram
prices=[50,60,45,70,120,90,85,40,30,200]

plt.hist(prices,bins=5)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Prices Distribution")
plt.show()
```

```python
#Box plot
plt.boxplot(prices)
plt.title("Price Boxplot")
plt.show()
```

```python
#Customazing styles
plt.style.use("ggplot") #Other options: 'seaborn','fivethrityeight', 'classic'"Py libraries.ipynb"

x=[1,2,3,4,5]
y=[2,4,6,8,10]

plt.plot(x,y)
plt.title("Styled line graph")
plt.show()
```

```python
#Annotating graphs
x=[1,2,3,4,5]
y=[2,4,6,8,10]

plt.plot(x,y)
plt.title("Annotated plot")

plt.annotate("Important point", xy=(3,6),xytext=(4,7), arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
#Multiple plot in one figure 
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y, label="Line")
plt.scatter(x,y, color="red", label="points" )
plt.title("Line+scatter")
plt.legend()
plt.show()
```

```python
#Color & colormaps
data=np.random.rand(5,5)
plt.imshow(data,cmap='viridis')#other example: 'plasma','inferno','coolwarm'"Py libraries.ipynb"
plt.colorbar() #show scale
plt.show()
```

```python
#Everything combined together
x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[1,3,5,7,9]

fig, ax=plt.subplots(figsize=(8,5))
ax.plot(x,y1, color="blue", marker="o",label="Even")
ax.plot(x,y2, color="green", marker="s", label="odd")
ax.set_title("Combined  Line plot")
ax.set_xlabel("X values")
ax.set_ylabel("Y values")
ax.grid(True, linestyle='--', alpha=0.5)
ax.legend()
plt.savefig("combined_plot.png", dpi=300, bbox_inches='tight')
plt.show()
```
